```python
data=[1,2,3,4,6,8,10,12]
fig,ax=plt.subplots(figsize=(5,5))
ax.set_xlabel('No of Elements',fontdict={'fontsize':10,'color':'red'})
ax.set_ylabel('Y - Values',fontdict={'fontsize':10,'color':'green'})
ax.grid(True)
ax.set_title("Basics of Vistualization",loc='right')
ax.plot(data,marker='o',markersize=5,markeredgecolor='red',markerfacecolor='cyan')
```

```python
data=[1,2,3,4,6,8,10,12]
fig,ax=plt.subplots(figsize=(5,5))
ax.set_xlabel('No of Elements',fontdict={'fontsize':10,'color':'red'})
ax.set_ylabel('Y - Values',fontdict={'fontsize':10,'color':'green'})
ax.grid(True)
ax.set_title("Basics of Vistualization",loc='right')
ax.plot(data,marker='X',markersize=15,markeredgecolor='red',markerfacecolor='cyan')
```

```python
data1=[1,2,3,4,6,8,10,12]
data2=[22,23,34,55,7,9,11,13]
fig,ax=plt.subplots(figsize=(10,5))
ax.plot(data1,marker='o',markersize=5,markeredgecolor='red',markerfacecolor='cyan')
ax.plot(data2,marker='X',markersize=5,markeredgecolor='red',markerfacecolor='cyan',
        linestyle='-',color='green',linewidth=2,alpha=0.5)
```

```python
data1=[1,2,3,4,6,8,10,12]
data2=[22,23,34,55,7,9,11,13]
fig,(ax1,ax2)=plt.subplots(figsize=(10,5),ncols=2)
ax1.plot(data1,marker='o',markersize=5,markeredgecolor='red',markerfacecolor='cyan')
ax2.plot(data2,marker='X',markersize=5,markeredgecolor='red',markerfacecolor='cyan',
        linestyle='--',color='green',linewidth=2,alpha=0.5)
```
