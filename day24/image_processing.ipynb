```python
from PIL import Image,ImageDraw,ImageFont,ImageFilter
import numpy as np
```

```python
#Creating a 'red' image of size 256x256 pixels
red_img=Image.new('RGB',(256,256),color='red')
red_img
```

```python
np.array(red_img)
#All color are made of 3 color that are red,green,blue and any other color is made of mixture of these three color. 
#In this column[0] is for red, column[1] is for blue, column[2] is for green
```

```python
circle=Image.new('RGBA',(256,256),(0,0,0,0))  #(256,256) size of rectangle and RGBA is used for making a image trasparent
draw=ImageDraw.Draw(circle)  #Draw is a class  of ImageDraw module of PIL it is used to draw figure
draw.circle((128,128),64,fill='red')  #64 is the radius of the circle and (128,128) is the center of the circle
draw.circle((128,128),50,fill='green') #50 is the  radius of the circle and (128,128) is the center of the circle
circle
```

```python
circle.save('my_circle.png')
```

#Image Manipulation

```python
img3=img2.resize((256,256))
img3
```

```python
img3.rotate(30,fillcolor='white')
```

```python
img3.rotate(30,fillcolor='white',expand=True)  #Remove all the cutting  lines
```

```python
img3.transpose(Image.Transpose.FLIP_LEFT_RIGHT)
```

```python
img4=Image.new('RGB',(512,512))
img4.paste(img3,(0,0))
img4.paste(img3.transpose(Image.Transpose.FLIP_LEFT_RIGHT),(256,0))
img4.paste(img3.rotate(45,fillcolor='White'),(0,256))
img4.paste(img3.rotate(-45,fillcolor='White'),(256,256))
img4.save("Formula1_mix.png")
```

### Filtering

```python
b=b.resize((300,300))
blur=b.filter(ImageFilter.BLUR)
outline=b.filter(ImageFilter.CONTOUR)
emboss=b.filter(ImageFilter.EMBOSS)
light=b.filter(ImageFilter.MaxFilter())
dark=b.filter(ImageFilter.MinFilter())
sharpen=b.filter(ImageFilter.SHARPEN)
blur.show()
outline.show()
emboss.show()
light.show()
dark.show()
sharpen.show()
```

adding text to image

```python
circle=Image.new('RGBA',(256,256),(0,0,0,0)) 
draw=ImageDraw.Draw(circle)  
draw.circle((128,128),64,fill='red')  
draw.circle((128,128),50,fill='green') 
font=ImageFont.load_default(size=50)
draw.text((128,128),'Hi!',font=font,fill='white',anchor='mm')
#draw.text((128,128+100),'Hello',font=font,fill='white',anchor='mm')
circle.save("my_circle2.png")
circle
```
